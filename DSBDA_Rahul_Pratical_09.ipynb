```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
import seaborn as sns

df = sns.load_dataset("titanic")
```

```python
print("First 5 rows of Titanic dataset:")
print(df.head())
```

```python
# Normalize column names
cols_lower = [col.lower() for col in df.columns]

age_col = df.columns[cols_lower.index('age')]
sex_col = df.columns[cols_lower.index('sex')]
surv_col = df.columns[cols_lower.index('survived')]
```

```python
# Check missing values
print("\nMissing values in Age before filling:", df[age_col].isnull().sum())
```

```python
# Fill missing age
df[age_col].fillna(df[age_col].median(), inplace=True)
```

```python
print("Missing values in Age after filling:", df[age_col].isnull().sum())
```

```python
# Convert to category
df[sex_col] = df[sex_col].astype('category')
df[surv_col] = df[surv_col].astype('category')
```

```python
# Boxplot
plt.figure(figsize=(10,6))

sns.boxplot(
    x=sex_col,
    y=age_col,
    hue=surv_col,
    data=df,
    palette="Set2"
)
```

```python
plt.title("Distribution of Age by Gender and Survival")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```
