```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
df = pd.read_csv(url)
```

```python
print("First 5 rows of the Iris dataset:")
print(df.head())
```

```python
grouped_stats = df.groupby('species')[['sepal_length', 'sepal_width']].describe()

print("\nSummary statistics of sepal dimensions grouped by species:")
print(grouped_stats)
```

```python
sepal_length_lists = df.groupby('species')['sepal_length'].apply(list)

print("\nSepal length values for each species:")
print(sepal_length_lists)
```

```python
def species_stats(df, species_name):
    species_data = df[df['species'] == species_name][['sepal_length', 'sepal_width']]

    print(f"\nBasic statistics for {species_name}:")

    print("\nMean:")
    print(species_data.mean())

    print("\nStandard Deviation:")
    print(species_data.std())

    print("\n25th percentile:")
    print(species_data.quantile(0.25))

    print("\n50th percentile (Median):")
    print(species_data.quantile(0.50))

    print("\n75th percentile:")
    print(species_data.quantile(0.75))
```

```python
species_list = df['species'].unique()

for species in species_list:
    species_stats(df, species)
```

```python
numeric_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for col in numeric_cols:
    plt.figure()
    sns.boxplot(x='species', y=col, data=df)
    plt.title(f"Boxplot of {col} by Species")
    plt.show()
```
