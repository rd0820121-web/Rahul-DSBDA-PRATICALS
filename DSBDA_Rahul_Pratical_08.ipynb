```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import seaborn as sns

df = sns.load_dataset("titanic")
```

```python
# Show first rows
print("First 5 rows of Titanic dataset:")
print(df.head())
```

```python
# Normalize column names
cols_lower = [col.lower() for col in df.columns]



age_col = df.columns[cols_lower.index('age')]
fare_col = df.columns[cols_lower.index('fare')]
sibsp_col = df.columns[cols_lower.index('sibsp')]
parch_col = df.columns[cols_lower.index('parch')]
surv_col = df.columns[cols_lower.index('survived')]
```

```python
# Handle missing values
print("\nMissing values in Age before filling:", df[age_col].isnull().sum())
df[age_col].fillna(df[age_col].median(), inplace=True)
print("Missing values in Age after filling:", df[age_col].isnull().sum())
```

```python
# Convert survival column
df[surv_col] = df[surv_col].astype("category")
```

```python
# Numeric columns
numeric_cols = [age_col, fare_col, sibsp_col, parch_col]
```

```python
# Pairplot
sns.pairplot(df[numeric_cols + [surv_col]], hue=surv_col, palette="Set1")
plt.suptitle("Pairplot of Titanic Dataset", y=1.02)
plt.show()
```

```python
# Histogram of Fare
plt.figure(figsize=(10,6))
sns.histplot(df[fare_col], bins=50, kde=True, color="skyblue")
```

```python
plt.title("Distribution of Titanic Ticket Fare")
plt.xlabel("Fare")
plt.ylabel("Number of Passengers")
plt.xlim(0, 300)
plt.show()
```
