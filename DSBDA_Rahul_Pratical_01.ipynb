```python
# Label Encoding (Alternative Method)

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df_label = df.copy()
for col in df_label.columns:
    df_label[col] = le.fit_transform(df_label[col])

print(df_label.head())
```
