```python
def myToken(text):
    ans = []
    curr = ""

    for ch in text:
        if ch == " ":
            if curr != "":
                ans.append(curr)
                curr = ""
        else:
            curr += ch

    if curr != "":
        ans.append(curr)

    print(ans)
```

```python
# convert text into word_token with these taga

from nltk.tokenize import word_tokenize
from nltk import pos_tag

def pos_tagg(text):
    word_tokens = word_tokenize(text)
    return pos_tag(word_tokens)

print(pos_tagg("Are you afraid of something?"))
```

```python
from nltk.text import TextCollection
from nltk.tokenize import word_tokenize

# Step 1: sentences
sentences = [
    "this is sentence one",
    "this is sentence two",
    "this is sentence three"
]

# Step 2: tokenize sentences
tokenized_sentences = [word_tokenize(sent) for sent in sentences]

print(tokenized_sentences)

# Step 3: build corpus
corpus = TextCollection(tokenized_sentences)

print(corpus)

# Step 4: TF (term frequency)
tf_value = corpus.tf("one", tokenized_sentences[0])
print("TF:", tf_value)

# Step 5: IDF (inverse document frequency)
idf_value = corpus.idf("one")
print("IDF:", idf_value)

# Step 6: TF-IDF
tfidf_value = corpus.tf_idf("one", tokenized_sentences[0])
print("TF-IDF:", tfidf_value)
```
