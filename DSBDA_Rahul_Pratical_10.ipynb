```python
import string
import math
from collections import Counter
```

```python
doc = """
Natural Language Processing (NLP) is a sub-field of artificial intelligence.
It focuses on the interaction between computers and humans through natural language.
It includes tasks such as text analysis, sentiment analysis, and machine translation.
"""
```

```python
# Preprocessing Function
def preprocess(text):
    text = text.lower()

    # remove punctuation
    text = text.translate(str.maketrans("", "", string.punctuation))

    tokens = text.split()

    stop_words = {
        'the','is','a','an','of','and','to','in','that','it','with',
        'such','through','this','by','are','on','as','for'
    }

    tokens = [word for word in tokens if word not in stop_words]

    return tokens
```

```python
# Tokenization
tokens = preprocess(doc)

print("Tokens after preprocessing:\n", tokens)
```

```python
# Simple Stemming
def simple_stem(word):
    suffixes = ['ing', 'ly', 'ed', 'es', 's', 'tion']

    for suf in suffixes:
        if word.endswith(suf) and len(word) > len(suf) + 2:
            return word[:-len(suf)]

    return word

stemmed_tokens = [simple_stem(word) for word in tokens]

print("\nStemmed Tokens:\n", stemmed_tokens)
```

```python
# Simple Lemmatization (Dictionary based)
lemma_dict = {
    "tasks": "task",
    "analyses": "analysis",
    "analysis": "analysis",
    "computers": "computer",
    "humans": "human",
    "translation": "translation",
    "processing": "process"
}

lemmatized_tokens = [lemma_dict.get(word, word) for word in stemmed_tokens]

print("\nLemmatized Tokens:\n", lemmatized_tokens)
```

```python
# Term Frequency (TF)
tf_dict = Counter(lemmatized_tokens)

print("\nTerm Frequency (TF):\n", dict(tf_dict))
```

```python
# Corpus for IDF
corpus = [
    doc,
    "Machine learning and NLP are important for AI and text analytics.",
    "Text processing involves tokenization stemming and lemmatization."
]
```

```python
# Preprocess Corpus

def preprocess_corpus(text):
    text = text.lower()
    text = text.translate(str.maketrans("", "", string.punctuation))
    tokens = text.split()

    stop_words = {
        'the','is','a','an','of','and','to','in','that','it','with',
        'such','through','this','by','are','on','as','for'
    }

    return set([word for word in tokens if word not in stop_words])

corpus_tokens = [preprocess_corpus(d) for d in corpus]
```

```python
# IDF Calculation

idf_dict = {}

N = len(corpus)

for word in tf_dict.keys():
    df_count = sum(1 for doc_set in corpus_tokens if word in doc_set)
    idf_dict[word] = math.log((N + 1) / (1 + df_count)) + 1

print("\nInverse Document Frequency (IDF):\n", idf_dict)
```

```python
# TF-IDF Calculation

tfidf_dict = {
    word: tf_dict[word] * idf_dict[word]
    for word in tf_dict.keys()
}

print("\nTF-IDF Representation:\n", tfidf_dict)
```
